```python
df = pd.read_csv('Fusion.csv',encoding='utf-8')
```

# SET 'ADDED_BY' NUMERIC VALUES TO USERNAME

```python
df.loc[df['added_by'] == '1279337455','added_by'] = 'Ulises'
df.loc[df['added_by'] == '2252aha4eok73iglx4byvzb3a', 'added_by'] = 'David' 
df.loc[df['added_by'] == 'eduardo_f04', 'added_by'] = 'Edi'
```

```python
df[['added_by','duration_sec']].groupby('added_by').mean().plot(kind = 'bar')
```
